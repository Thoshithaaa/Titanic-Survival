```python
import re
deck = {"A": 1, "B": 2, "C": 3, "D": 4, "E": 5, "F": 6, "G": 7, "U": 8}
data = [train_df, test_df]
for dataset in data:
  dataset['Cabin'] = dataset['Cabin'].fillna("U0")
  dataset['Deck'] = dataset['Cabin'].map(lambda x: re.compile("([a-zA-Z]+)").search(x).group())
  dataset['Deck'] = dataset['Deck'].map(deck)
  dataset['Deck'] = dataset['Deck'].fillna(0)
  dataset['Deck'] = dataset['Deck'].astype(int)
```

**Hyperparameter Tuning**

```python
def clf_performance(classifier, model_name):
    print(model_name)
    print('Best Score: ' + str(classifier.best_score_))
    print('Best Parameters: ' + str(classifier.best_params_))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Using only the first 80 samples of Y_train
Y_train_subset = Y_train[:80]

# Define the RandomForestClassifier
rf = RandomForestClassifier(random_state=1)

# Define the reduced parameter grid
param_grid =  {'n_estimators': [400, 500],
               'criterion': ['gini', 'entropy'],
               'bootstrap': [True],
               'max_depth': [15, 20],
               'max_features': ['auto', 'sqrt'],
               'min_samples_leaf': [2, 3],
               'min_samples_split': [2, 3]}

# Create GridSearchCV with reduced grid and 3-fold cross-validation
clf_rf = GridSearchCV(rf, param_grid=param_grid, cv=3, verbose=True, n_jobs=-1)

# Fit the model using only the first 80 samples of X_train and corresponding Y_train_subset
best_clf_rf = clf_rf.fit(X_train[:80], Y_train_subset)

# Print the best parameters found
print("Best parameters found: ", best_clf_rf.best_params_)

# Assuming you have defined clf_performance
# Print score
clf_performance(best_clf_rf, 'Random Forest')
```
